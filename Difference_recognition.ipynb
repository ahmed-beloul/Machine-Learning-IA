```python
import cv2
import imutils 
import matplotlib.pyplot as plt
```

```python
#we read the two images 
perfect = cv2.imread("carpaint.jpg")
defect = cv2.imread("cardefaut5.jpg")
#resize the images

perfect = imutils.resize(perfect, width = 500)
defect = imutils.resize(defect, width = 500)
plt.imshow(perfect)
plt.show()
plt.imshow(defect)
plt.show()

```

```python
sub=cv2.subtract(defect, perfect)
plt.imshow(sub)
plt.show()
```

```python
#converting the image in grayscale
subgray = cv2.cvtColor(sub, cv2.COLOR_BGR2GRAY)
plt.imshow(subgray)
plt.show()  
#increase the size of the default 
for i in range(0, 3):
    dil = cv2.dilate(subgray.copy(), None, iterations= i+ 1)
plt.imshow(dil)
plt.show()
```

```python
#threshold of the image 
(T, thresh) = cv2.threshold(dil, 10, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
  
#find the contours 
contours = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)
```

```python
for c in contours:
  (x, y, w, h) = cv2.boundingRect(c)
  cv2.rectangle(defect, (x, y), (x + w, y + h), (0, 255, 0), 2)
  

cv2.imwrite("new.png", defect)
plt.imshow(cv2.imread("new.png"))

```
